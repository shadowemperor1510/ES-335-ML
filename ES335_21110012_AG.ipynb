Question 1:
Consider two matrices:

Matrix (A) of size (n * k).
Matrix (B) of size (k * m).
The product matrix (C = A * B) has dimensions (n * m), and each element in (C) is obtained by multiplying a row of (A) with a column of (B). In this process:

For a given element at position ((i, j)) in (A), it is multiplied (k) times with the corresponding element in the (j)th column of (B).
The total number of multiplications for each element in the product matrix is (k).
Therefore, with (n) rows and (m) columns in the product matrix, the total number of multiplications involved in obtaining (C) is given by: k * m * n

Each element of product is formed by adding  k numbers. hence, there will be k-1 plus signs. This is true for each element of product matrix, so the total addition signs will be: (k-1) * m * n

```python
# Question 2
import numpy as np
import time

def matrix_multiply(a, b):
    n = len(a)
    k = len(a[0])
    m = len(b[0])

    result = [[0] * m for _ in range(n)]

    for i in range(n):
        for j in range(m):
            for l in range(k):
                result[i][j] += a[i][l] * b[l][j]

    return result

def np_ans(a,b):
    return np.dot(a,b)

n = int(input())
k = int(input())
m = int(input())
a=[]
b=[]
for i in range(n):
    c=[]
    for j in range(k):
        num=int(input())
        c.append(num)
    a.append(c)

for i in range(k):
    c=[]
    for j in range(m):
        num=int(input())
        c.append(num)
    b.append(c)

result_list = matrix_multiply(a, b)

# Display the result
print("Matrix multiplication result using list of lists:")
for row in result_list:
    print(row)

a_np = np.array(a)
b_np = np.array(b)
result_np= np_ans(a_np,b_np)

print("\nMatrix multiplication result using NumPy:")
print(result_np)

# Timing for list of lists
start_time = time.time()
result_list = matrix_multiply(a, b)
print("List of lists timing:", time.time() - start_time)

# Timing for NumPy
start_time = time.time()
result_np = np_ans(a_np, b_np)
print("NumPy timing:", time.time() - start_time)

# Numpy is faster because it is specifically created for numerical operations. Its code is compiled in C. It optimizes the calculations. Whereas, list of lists multiplication is done for 1 element at a time.
```

Question 3:

Largest element is found in the one pass, 2nd highest is found in 2 passes. So, the kth largest element will be found in k passes. One pass means traversing the entire list. So, to find the median, the T.C. would be O(n^2).
A better method would be to sort the array using any sorting technique like merge sort or quick sort or the inbuilt sort in python with a better T.C. than O(n^2).

```python
# Question 3
# Since the T.C. is O(n^2), I am using bubble sort to compare time.

import numpy as np
import time

def bub(a):
    n = len(a)
    for i in range(n):
        for j in range(0, n-i-1):
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]
    

n=int(input())
a=[int(input()) for _ in range(n)]
temp = a

st=time.time()
bub(a)
print("Time for bubble sort = ", time.time()-st)

st=time.time()
temp.sort()
print("Time for inbuilt = ", time.time()-st)

a=np.array(a)
st=time.time()
ans=np.median(a)
print("Time for numpy = ", time.time()-st)

# numpy and inbuilt are faster.
```
